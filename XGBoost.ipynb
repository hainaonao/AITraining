**Toán học đằng sau XgBoost:**
Trước khi bắt đầu với toán học về Gradient Boosting, đây là một ví dụ đơn giản về một CART (Cây quyết định phân loại và hồi quy) phân loại liệu ai đó sẽ thích một trò chơi máy tính giả định X. Dự đoán điểm của mỗi cây quyết định cá nhân sau đó được tổng hợp để có được $(\hat{y}_i)$. Một điểm quan trọng trong ví dụ này là hai cây cố gắng bổ sung cho nhau.

Hàm mục tiêu cho mô hình trên có thể được viết dưới dạng:

$  \hat{y}_{i} =  \sum_{k=1}^{K} f_k(x_i), f_k \epsilon \mathcal{F}  $

Trong đó, \(K\) là số cây, \(f\) là không gian chức năng của \(F\), \(F\) là tập hợp các CART có thể có. Hàm mục tiêu cho mô hình trên có thể được định nghĩa là:

$ obj(\theta) = \sum_{i}^{n} l(y_{i}, \hat{y}_{i}) + \sum_{k=1}^K \Omega(f_{k}) $

Nơi, thành phần đầu tiên là hàm mất mát và thành phần thứ hai là tham số điều chuẩn. Thay vì học cây một cách tức thì, làm cho việc tối ưu hóa khó khăn hơn, chúng ta áp dụng chiến lược cộng dồn, giảm thiểu mất mát đã học và thêm một cây mới có thể tóm tắt như sau:

$ \hat{y}_i^{(0)} =0\\ \hat{y}_i^{(1)} =  f_1(x_i) = \hat{y}_i^{(0)} + f_1(x_i)\\ \hat{y}_i^{(2)} = f_1(x_i) + f_2(x_i)= \hat{y}_i^{(1)} + f_2(x_i)\\ ....\\ \hat{y}_i^{(t)} = \sum_{k=1}^t f_k(x_i)= \hat{y}_i^{(t-1)} + f_t(x_i) $

Hàm mục tiêu của mô hình trên có thể được định nghĩa là:

$ obj^{(t)}  = \sum_{i=1}^{n} l(y_i, \hat{y}_{i}^{(t)}) + \sum_{i=1}^t\Omega(f_i) \\           = \sum_{i=1}^{n} l(y_i, \hat{y}_{i}^{(t-1)} + f_{t}(x_i)) + \Omega(f_t) + hằng số $

Nơi, thành phần đầu tiên là hàm mất mát và thành phần thứ hai là tham số điều chuẩn. Bây giờ, chúng ta áp dụng phương pháp mở rộng chuỗi Taylor đến bậc hai:

$ obj^{(t)} = \sum_{i=1}^{n} [l(y_i, \hat{y}_i^{(t-1)}) + g_i f_t(x_i) + \frac{1}{2} h_{i} f_{t}^2(x_i)] + \Omega(f_t) + hằng số $

Ở đây, $g_i$ và $h_i$ có thể được định nghĩa như sau:

$ g_i = \partial_{\hat{y}_i^{(t-1)}} l(y_i, \hat{y}_i^{(t-1)}) $
$ h_i = \partial_{\hat{y}_i^{(t-1)}}^2 l(y_i, \hat{y}_i^{(t-1)}) $

Giảm đơn giản và loại bỏ hằng số, chúng ta có:

$ \sum_{i=1}^n [g_{i} f_{t}(x_i) + \frac{1}{2} h_{i} f_{t}^2(x_i)] + \Omega(f_t) $

Bây giờ, chúng ta định nghĩa thuật ngữ điều chuẩn, nhưng trước hết chúng ta cần định nghĩa mô hình:
Ở đây, $f_t(x) = w_{q(x)}, w \in R^{T}, q:R^d\rightarrow \{1,2,...,T\}$ nghĩa là \(w\) là vector điểm số trên các lá của cây, \(q\) là hàm gán mỗi điểm dữ liệu vào lá tương ứng, và \(T\) là số lá của cây. Thuật ngữ điều chuẩn được định nghĩa như sau:

$\Omega(f) = \gamma T + \frac{1}{2}\lambda \sum_{j=1}^T w_j^2$

Bây giờ, hàm mục tiêu của chúng ta trở nên:

$obj^{(t)} \approx \sum_{i=1}^n [g_i w_{q(x_i)} + \frac{1}{2} h_i w_{q(x_i)}^2] + \gamma T + \frac{1}{2}\lambda \sum_{j=1}^T w_j^2\]
\[= \sum^T_{j=1} \left[ \left(\sum_{i\in I_j} g_i\right) w_j + \frac{1}{2} \left(\sum_{i\in I_j} h_i + \lambda\right) w_j^2 \right] + \gamma T$

Tiếp theo, chúng ta đơn giản hóa biểu thức trên:

$obj^{(t)} = \sum^T_{j=1} \left[ G_jw_j + \frac{1}{2} \left(H_j+\lambda\right) w_j^2 \right] +\gamma T$

Trong đó:

$G_j = \sum_{i\in I_j} g_i\\ H_j = \sum_{i\in I_j} h_i$

Trong phương trình này, $w_j$ là độc lập với nhau, $w_j$ tốt nhất cho một cấu trúc $q(x)$ đã cho và giảm thiểu mục tiêu tốt nhất mà chúng ta có thể đạt được là:

$w_j^{\ast} = -\frac{G_j}{H_j+\lambda}\\ \text{obj}^{\ast} = -\frac{1}{2} \sum_{j=1}^T \frac{G_j^2}{H_j+\lambda} + \gamma T$

Ở đây, $\gamma$ là tham số cắt tỉa, tức là mức giảm thông tin tối thiểu để thực hiện phân nhánh.

Bây giờ, chúng ta cố gắng đo lường xem cây tốt như thế nào, chúng ta không thể tối ưu trực tiếp cây, chúng ta sẽ cố gắng tối ưu một cấp của cây mỗi lần. Cụ thể, chúng ta cố gắng phân chia một lá thành hai lá, và điểm số mà nó đạt được là:

$Gain = \frac{1}{2} \left[\frac{G_L^2}{H_L+\lambda}+\frac{G_R^2}{H_R+\lambda}-\frac{(G_L+G_R)^2}{H_L+H_R+\lambda}\right] - \gamma$

```python
# Phần 1. GBM
# Data: Loan Data (Upload Github)

import pandas as pd
import numpy as np


```

```python
# Doc du lieu tu file
df = pd.read_csv("/content/loan_data.csv")
```

```python
df
```

```python
# Xử lý dữ liệu
df.dropna(inplace=True)
df.drop(columns=['Loan_ID'], inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

is_Category = df.dtypes == object

category_column_list = df.columns[is_Category].tolist()

df[category_column_list] = df[category_column_list].apply(lambda col: le.fit_transform(col))
```

```python
df
```

```python
# Chia dữ liệu train, test

from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size = 0.2, random_state=42)

x_train = train.drop(columns=['Loan_Status'])
y_train = train['Loan_Status']

x_test = test.drop(columns=['Loan_Status'])
y_test = test['Loan_Status']
```

```python
# Train model

from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier(learning_rate=0.01, random_state=42)
model.fit(x_train, y_train)

```

```python
# Tinh toan acc tren test

acc = model.score(x_test, y_test)
print(acc*100)
```

```python
# XgBoost

import xgboost as xgb

model_xgb = xgb.XGBClassifier(learning_rate =0.01, random_state=42, n_estimators = 100)
model_xgb.fit(x_train, y_train)
```

```python
acc = model_xgb.score(x_test, y_test)
print(acc*100)
```
